# 3- model 3 GRU

```python
input = layers.Input((1), dtype =tf.string)
x = text_vectorizer(input)
x = embidding(x)
#x = layers.LSTM(64, return_sequences=True) (x)
x = layers.GRU(64)(x)
#x = layers.Dense(64, activation="relu")(x)
output = layers.Dense(1, activation = "sigmoid") (x)
model_3 = tf.keras.Model(input, output)
```

```python
model_3.compile(
    loss = "binary_crossentropy",
    optimizer = tf.keras.optimizers.Adam(),
    metrics= "accuracy"
)
history = model_3.fit( train,train_labels, epochs = 10, validation_data =(vaild,vaild_labels)
                      ,
                      callbacks = [create_tensorboard_callback("tensorboard", "model_3 with LSTM model")])
```

# 4- Bidirectional

```python
input = layers.Input((1), dtype =tf.string)
x = text_vectorizer(input)
x = embidding(x)
#x = layers.LSTM(64, return_sequences=True) (x)
x = layers.Bidirectional(layers.LSTM(64))(x)

#x = layers.Dense(64, activation="relu")(x)
output = layers.Dense(1, activation = "sigmoid") (x)
model_3 = tf.keras.Model(input, output)
```

```python
model_3.compile(
    loss = "binary_crossentropy",
    optimizer = tf.keras.optimizers.Adam(),
    metrics= "accuracy"
)
history = model_3.fit( train,train_labels, epochs = 10, validation_data =(vaild,vaild_labels)
                      ,
                      callbacks = [create_tensorboard_callback("tensorboard", "model_3 with LSTM model")])
```

# 5- conv1D

```python
input = layers.Input((1), dtype =tf.string)
x = text_vectorizer(input)
x = embidding(x)
#x = layers.LSTM(64, return_sequences=True) (x)
x = layers.Conv1D(filters= 64, kernel_size=5 ,padding = "same", activation = "relu")(x)
x = layers.GlobalMaxPool1D()(x)

#x = layers.Dense(64, activation="relu")(x)
output = layers.Dense(1, activation = "sigmoid") (x)
model_5 = tf.keras.Model(input, output)
```

```python
model_5.compile(
    loss = "binary_crossentropy",
    optimizer = tf.keras.optimizers.Adam(),
    metrics= "accuracy"
)
history = model_5.fit( train,train_labels, epochs = 10, validation_data =(vaild,vaild_labels)
                      ,
                      callbacks = [create_tensorboard_callback("tensorboard", "model_3 with LSTM model")])
```

```python
embd = hub.KerasLayer("https://tfhub.dev/google/universal-sentence-encoder/4",
                      trainable = False,
                      input_shape = [],
                      dtype = tf.string)


model_6 = tf.keras.Sequential([
    embd,
    layers.Dense(64, activation="relu"),
    layers.Dense(1, activation = "sigmoid") 
])
```
